```python
import pandas as pd
import matplotlib.pyplot as plt

pnl = pd.read_csv('pnl.csv')
history = pd.read_csv('price_history.csv')
yes_orders = pd.read_csv('yes_orders.csv')
no_orders = pd.read_csv('no_orders.csv')

step = len(pnl['timestamp']) // 20

plt.figure(1)
plt.plot(pnl['timestamp'], pnl['PnL'])
plt.xlabel('Time')
plt.ylabel('PnL')
plt.xticks(ticks=range(0, len(pnl['timestamp']), step), labels=pnl['timestamp'][::step], rotation=90)
plt.title('PnL over Time')
plt.tight_layout()
plt.show()

plt.figure(2)
plt.plot(history['timestamp'], history['best_a'], label='Best Offer')
plt.plot(history['timestamp'], history['best_b'], label='Best Bid')
plt.scatter(yes_orders['timestamp'], yes_orders['price'], label='Yes Orders', color='green', marker='o')
plt.scatter(no_orders['timestamp'], no_orders['price'], label='No Orders', color='red', marker='o')
plt.xlabel('Time')
plt.ylabel('Price')
plt.xticks(ticks=range(0, len(history['timestamp']), step), labels=history['timestamp'][::step], rotation=90)
plt.ylim(0,100)
plt.title('Price History with Orders')
plt.legend()
plt.tight_layout()
plt.show()
```
